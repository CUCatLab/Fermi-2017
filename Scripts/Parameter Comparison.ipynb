```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
###################
### Import Data ###
###################

Folder = 'C:\\Users\\larue\\OneDrive - Chapman University\\Research\\Projects\\X-Ray Femtochemistry\\Elettra & Fermi Lightsources\\2017\\Data Analysis\\'
File = 'Parameters.xlsx'
TimeResolution = 50

H21 = pd.read_excel(Folder+File,sheet_name='2-1 Hor '+str(TimeResolution))
V21 = pd.read_excel(Folder+File,sheet_name='2-1 Ver '+str(TimeResolution))
H12 = pd.read_excel(Folder+File,sheet_name='1-2 Hor '+str(TimeResolution))
NumPeaks_H21 = int((H21.shape[1]-4)/3)
NumPeaks_V21 = int((H21.shape[1]-4)/3)
NumPeaks_H12 = int((H21.shape[1]-4)/3)
```

```python
###################
### Plot Trends ###
###################

# Negative delay
plt.plot(H21['Delay'], H21['G1_amp'], '.:', label='2-1 Hor')
plt.plot(V21['Delay'], V21['G1_amp'], '.:', label='2-1 Ver')
plt.plot(H12['Delay'], H12['G1_amp'], '.:', label='1-2 Hor')
plt.legend(), plt.xlabel('Delay (fs)'), plt.ylabel('Amplitude')
plt.title('Gaussian 1: Negative delay')
plt.show()

# Gas phase
plt.plot(H21['Delay'], H21['G2_amp'], '.:', label='2-1 Hor')
plt.plot(V21['Delay'], V21['G2_amp'], '.:', label='2-1 Ver')
plt.plot(H12['Delay'], H12['G2_amp'], '.:', label='1-2 Hor')
plt.legend(), plt.xlabel('Delay (fs)'), plt.ylabel('Amplitude')
plt.title('Gaussian 2: Gas Phase')
plt.show()

# Unknown
plt.plot(H21['Delay'], H21['G7_amp'], '.:', label='2-1 Hor')
plt.plot(V21['Delay'], V21['G7_amp'], '.:', label='2-1 Ver')
plt.plot(H12['Delay'], H12['G7_amp'], '.:', label='1-2 Hor')
plt.legend(), plt.xlabel('Delay (fs)'), plt.ylabel('Amplitude')
plt.title('Gaussian 7: Unknown')
plt.show()
```

```python
##################################
### Negative and Unknown Ratio ###
##################################

Ratio_H21 = H21['G1_amp'] / H21['G7_amp']
Ratio_V21 = V21['G1_amp'] / V21['G7_amp']
Ratio_H12 = H12['G1_amp'] / H12['G7_amp']
plt.plot(H21['Delay'], Ratio_H21, '.:', label='2-1 Hor')
plt.plot(V21['Delay'], Ratio_V21, '.:', label='2-1 Ver')
plt.plot(H12['Delay'], Ratio_H12, '.:', label='1-2 Hor')
plt.legend(), plt.xlabel('Delay (fs)'), plt.ylabel('Amplitude')
plt.title('Negative and Unknown Ratio')
plt.show()

####################################
### Negative and Gas Phase Ratio ###
####################################

Ratio_H21 = H21['G1_amp'] / H21['G2_amp']
Ratio_V21 = V21['G1_amp'] / V21['G2_amp']
Ratio_H12 = H12['G1_amp'] / H12['G2_amp']
plt.plot(H21['Delay'], Ratio_H21, '.:', label='2-1 Hor')
plt.plot(V21['Delay'], Ratio_V21, '.:', label='2-1 Ver')
plt.plot(H12['Delay'], Ratio_H12, '.:', label='1-2 Hor')
plt.legend(), plt.xlabel('Delay (fs)'), plt.ylabel('Amplitude')
plt.title('Negative and Gas Phase Ratio')
plt.show()

###################################
### Gas Phase and Unknown Ratio ###
###################################

Ratio_H21 = H21['G2_amp'] / H21['G7_amp']
Ratio_V21 = V21['G2_amp'] / V21['G7_amp']
Ratio_H12 = H12['G2_amp'] / H12['G7_amp']
plt.plot(H21['Delay'], Ratio_H21, '.:', label='2-1 Hor')
plt.plot(V21['Delay'], Ratio_V21, '.:', label='2-1 Ver')
plt.plot(H12['Delay'], Ratio_H12, '.:', label='1-2 Hor')
plt.legend(), plt.xlabel('Delay (fs)'), plt.ylabel('Amplitude')
plt.title('Gas Phase and Unknown Ratio')
plt.show()
```
