# Data binner
#### For FERMI data from Oct/Nov 2017 at DiProI beamline

## Info

This script works on the HDF5 files created by the Data collector (16_data_collection.ipynb) that end on '_col.h5' and are located in the 'combined' folder of each run to bin mass spec time of flight data.

# Load Packages

```python
import numpy as np
from matplotlib import pyplot as plt
import h5py
import sys
import time
import scipy.stats
import os.path
from importlib import reload
from pandas import DataFrame as df
import math
```

# Parameters

```python
##################
### Histograms ###
##################

# Number of bins
n_bin_counts  = 100

###############
### Filters ###
###############

# Mass spec count rate
filter_ms     = True
ms_low_thr    = -np.inf
ms_hi_thr     = np.inf

# LASER intensity
filter_laser  = True
laser_low_thr = 715
laser_hi_thr  = 730

###################
### Run(s) ########
###################

run_path = 'D:/FERMI 2017 2/CO_O_laserOnly/combined/XAS099_CO_O_02_col.h5'
```

# Bin Data

```python
%matplotlib inline

# Ignore Runtime Warnings for divide and invalid values
np.seterr(divide='ignore', invalid='ignore')

####################################
### --- Start of Main Script --- ###
####################################

# Check that data path exists
if not os.path.exists(run_path) :
    print(run_path)
    raise NameError('Check run_path! File does not exist!')
        
print('Loading data from:')
print(run_path)

# Define empty arrays to collect data in
laser_int_all        = []
ms_counts_all        = []
ms_tof_all           = []

#################
### Load Data ###
#################

# Loop over collected files

fh5 = h5py.File(run_path, 'r')

# LASER
laser_int = fh5['LASER/laser_int'][()]
laser_int_all.extend(laser_int)

# MS counts
ms_counts = fh5['/MASSSPEC/counts'][()]
ms_tof = fh5['/MASSSPEC/tof'][()]

################
### Bin Data ###
################

ms_hist, ms_bin_edges = np.histogram(ms_tof,bins=n_bin_counts)
ms_bins = list()
i = 0
while i < len(ms_hist) :
    ms_bins.append(np.mean((ms_bin_edges[i],ms_bin_edges[i+1])))
    i += 1
ms_hist = ms_hist / len(ms_counts)

plt.plot(ms_bins, ms_hist)
plt.ylabel('Number of Counts')
plt.xlabel('Time of Flight')
plt.show()

ToCopy = df(data=ms_hist,index=ms_bins)
ToCopy.to_clipboard()
```
